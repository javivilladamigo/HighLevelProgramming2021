```python
# 01ex - 2

#swapping values with tmp
def swap_function(x, y):
    z = x
    x = y
    y = z
    return x, y

#swapping values without tmp
def swap_function2(x, y):
    x, y = y, x
    return x, y
```

```python
# 01ex - 4

s = "Write a program that prints the numbers from 1 to 100. \
But for multiples of three print Hello instead of the number and for the multiples of five print World. \
For numbers which are multiples of both three and five print HelloWorld."

def counter(txt):
  count = {}
  for letter in txt:
    if letter == ' ' or letter == '\t':
      continue
    if letter in count:
      count[letter] += 1
    else:
      count[letter] = 1
  for k, v in count.items():
      print(f'Charcater {k} occurs {v} times')

counter(s)
```

```python
# 01ex - 6

#x = int(input("enter your number: "))

def square(num):
    return num ** 2

def cube(num):
    return num ** 3

def power_6th(num):
    return cube(square(num))

#print(power_6th(x))
```

```python
# 01ex - 7

# a. using a for loop
a = []

for x in range(1, 11):
    b = x ** 3
    a.append(b)

print(a)

# b. using a list comprehension
a = [x ** 3 for x in range(1, 11)]

print(a)
```

```python
# 01ex - 8

# way 1
res = tuple()
for x in range(1, 100):
    for y in range(x, 100):
        for z in range(1, 100):
            if x*x + y*y == z*z:
                res += ((x, y, z),)

print(res)

# way 2
res = [(a, b, c) for a in range(1, 100) for b in range(a, 100) for c in range(b, 100) if a**2+b**2 == c**2]

print(res)
```
