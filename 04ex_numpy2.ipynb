5\. Create a matrix that shows the 10 by 10 multiplication table.

 * Find the trace of the matrix
 * Extract the anti-diagonal matrix (this should be ```array([10, 18, 24, 28, 30, 30, 28, 24, 18, 10])```)
 * Extract the diagonal offset by 1 upwards (this should be ```array([ 2,  6, 12, 20, 30, 42, 56, 72, 90])```)

````python
a = []

for i in range(1,11):
    for j in range(1,11):
        a.append(i * j)

b = np.array(a)
c = b.reshape(10,10)
print("10 by 10 multiplication table")
print(c, "\n")

Diag = np.diag(c)
print("The diagonal of the matrix: {}".format(Diag))
print("The trace of the matrix: {}".format(sum(Diag)))
print("\n")

antiDiag = np.diag(c[::-1])
print("The anti-diagonal matrix: {}".format(antiDiag))

a2 = a
for k in antiDiag:
    a2.remove(k)

print(np.array(a2).reshape(10,9), "\n")

offsetDiag = np.diag(c,1)
print("Diagonal offset by 1 upwards: {}".format(offsetDiag))

a3 = a

for l in offsetDiag:
    a3.remove(l)

print(np.array(a3).reshape(9,9))
````
