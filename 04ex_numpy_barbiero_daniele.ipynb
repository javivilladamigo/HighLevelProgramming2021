2\. Find the outer product of the following vectors:

```python
u = np.array([1, 3, 5, 7])
v = np.array([2, 4, 6, 8])
```

Use different methods to do this:

   1. Using the function `outer` in numpy
   2. Using a nested `for` loop or a list comprehension
   3. Using numpy broadcasting operations

````python
u = np.array([1, 3, 5, 7])
v = np.array([2, 4, 6, 8])
z=np.zeros((u.shape[0], v.shape[0]))

print(np.outer(u,v))   #using outer function

for i in range(0,u.shape[0]): #using a nested loop
    for j in range(0,v.shape[0]):
        z[i,j]=(u[i]*v[j])
         
print(z)    

k=np.tile(u, (1,1)).T  #using the broadcasting
print(k*v)
````

```python
m = [[0. 0.1, 0.2 0.3 ... 0.9]
     [1.0 1.1 1.2 1.3 ... 1.9]
     ...
     [5.0 5.1 5.2 5.3 ... 5.9]]
```
After creating the matrix, set all entries with a decimal part $< 0.4$ to zero.

````python
tmp=np.fromfunction(lambda i, j: (i)+((j)/10), (6, 10))  #to create the matric
print(tmp)

for i in range(0,tmp.shape[0]):
    for j in range(0,tmp.shape[1]):
        if tmp[i,j]<0.4:
            tmp[i,j]=0     #to set all entries with a decimal part  <0.4  to zero

print('\n',tmp)
````

````python
m = np.fromfunction(lambda i, j: (i+1)*(j+1) ,(10, 10))
print('the multiplication matrix ')
print(m)
print('\nits antidiagonal:')
print(np.fliplr(m).diagonal())

z=np.zeros(m.shape[0])  #I add a row of zeros
new = np.vstack([z, m])
#print(new)
diag=new.diagonal()    #then I find the usual diagonal
diag = np.delete(diag, 0)   #and remove the term in excess
print('\noffset diagonal:')
print(diag)
````
